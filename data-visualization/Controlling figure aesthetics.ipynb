```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(sum(map(ord, "aesthetics")))
```

Define a function to plot some offset sin waves

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i*.5)*(7-i)*flip)
```

### Styles

Use matplot default style

```python
sinplot()
```

Use seaborn defaults style

```python
sns.set()
```

```python
sinplot()
```

There are 5 preset seaborn themes: `darkgrid`, `whitegrid`, `dark`, `white`, and `ticks`, the default is `darkgrid`

```python
sns.set_style('white')
sinplot()
```

```python
sns.set_style('whitegrid')
sinplot()
```

```python
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('ticks')
sinplot()
```

### Remove axes spines

```python
sinplot()
sns.despine() #
```

More settings of `despine`

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.violinplot(data=data)
sns.despine(offset=10, trim=True)
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True) #remove left
```

### Temporarily setting figure style

```python
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

Set all the axes styles which can be override

```python
sns.axes_style()
```

```python
sns.set_style('darkgrid', {'axes.facecolor': '.9', 'grid.linestyle':'--'})
sinplot()
```
