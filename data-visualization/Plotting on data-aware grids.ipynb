```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

sns.set(style='ticks')
np.random.seed(sum(map(ord, 'axis_grid')))
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(tips, col='time')
```

```python
g = sns.FacetGrid(tips, col='sex', hue='smoker')
g.map(plt.scatter, 'total_bill', 'tip', alpha=.7)
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col='day')
```

```python
g = sns.FacetGrid(tips, col='time')
g.map(plt.hist, 'tip')
```

```python
g = sns.FacetGrid(tips, col='sex', hue='smoker')
g.map(plt.scatter, 'total_bill', 'tip', alpha=.7)
g.add_legend()
```

```python
sns.factorplot(data=tips, x='total_bill', y='tip', 
               col='sex',hue='smoker',kind='strip' )
```

```python
g = sns.FacetGrid(tips, row='smoker', col='time', margin_titles=True)
g.map(sns.regplot, 'size', 'total_bill',  fit_reg=False, x_jitter=.1)
```

```python
g = sns.FacetGrid(tips, col='day', size=4, aspect=.5)
g.map(sns.barplot, 'sex', 'total_bill')
```

```python

titanic = titanic.assign(deck=titanic.deck.astype(object)).sort_values('deck')
```

```python
g = sns.FacetGrid(titanic, col='class', sharex=False, 
                 gridspec_kws={'width_ratios': [5, 3, 3]})
g.map(sns.boxplot, 'deck', 'age')
```

```python
ordered_days = tips.day.value_counts().index
```

```python
ordered_days
```

```python
tips.day.value_counts()
```

```python
g = sns.FacetGrid(tips, row='day',
                  row_order=ordered_days,
                 size=1.7, aspect=4)
g.map(sns.distplot, 'total_bill', hist=False, rug=True)
```

```python
pal = dict(Lunch='seagreen', Dinner='gray')
g = sns.FacetGrid(tips, hue='time', palette=pal, size=5)
g.map(plt.scatter, 'total_bill', 'tip',
      s=50,
      alpha=0.7, linewidth=.5, 
     edgecolor='white')
g.add_legend()
```

```python
g= sns.FacetGrid(tips, hue='sex', palette='Set1', size=5, 
                 hue_kws={'marker':['^', 'v']})
g.map(plt.scatter, 'total_bill', 'tip', s=100, 
     linewidth=.5, edgecolor='white')
```

```python
attend= sns.load_dataset('attention').query('subject <=12')
g = sns.FacetGrid(attend, col='subject', col_wrap=4, size=2, ylim=(0,10))
g.map(sns.pointplot, 'solutions', 'score', color='.3', ci=None)
```

```python
with sns.axes_style('white'):
    g = sns.FacetGrid(tips, row='sex', col='smoker',
                     margin_titles=True, size=2.5)
g.map(plt.scatter, 'total_bill', 'tip', color='#334455',
     edgecolor='white', lw=.5)
g.set_axis_labels('Total bill (US Dollars)', "tip")
g.set(xticks=[10,30,50], yticks=[2,6,10])
g.fig.subplots_adjust(wspace=.02, hspace=.02)
```

```python
iris = sns.load_dataset('iris')
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris, hue='species')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```

```python
g = sns.PairGrid(iris, vars=["sepal_length", "sepal_width"], hue="species")
g.map(plt.scatter);
```

```python
g = sns.PairGrid(iris)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_diag(sns.kdeplot, lw=3, legend=True);
```

```python
g = sns.pairplot(iris, hue="species", palette="Set1", diag_kind="kde", size=2.5)
```
