## Distributions of observations within categories

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
```

```python
tips['weekend'] = tips['day'].isin(['Sat', 'Sun'])
sns.boxplot(x='day', y='total_bill', hue='weekend', data=tips, dodge=True)
```

```python
sns.boxplot(x='day', y='total_bill', hue='weekend', data=tips, dodge=False)
```

## Violinplots

```python
sns.violinplot(x='total_bill', y='day', hue='time', data=tips)
```

```python
sns.violinplot(x='total_bill', y='day', hue='time', data=tips, bw=.1, scale='count', scale_hue=False)
```

```python
sns.violinplot(x='total_bill', y='day', hue='time', data=tips, bw=.1, scale='count', scale_hue=True)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True, 
              inner='stick', palette='Set3')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, inner=None)
sns.swarmplot(x='day', y='total_bill', data=tips, color='w', alpha=.5)
```

## Statistical estimation within categories

### Bar plots

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
sns.countplot(x='deck', data=titanic, palette='Greens_d')
```

```python
sns.countplot(y='deck', hue='class', data=titanic, palette="Greens_d")
```

### Point plots

```python
sns.pointplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
sns.pointplot(x='class', y='survived', hue='sex', data=titanic, 
             palette={'male':'g', 'female':'m'}, 
             markers=['^', 'o'],
             linestyles=['-', '--'])
```
