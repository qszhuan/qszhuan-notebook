## Prepare data

```python
from sklearn.datasets.samples_generator import make_blobs

centers = [
    [-2, 2],
    [2, 2],
    [0, 4]
]

x, y = make_blobs(n_samples = 60, centers=centers, random_state=0, cluster_std=0.60)
```

## Plotting

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(16,10), dpi=144)

c = np.array(centers)
plt.scatter(x[:, 0], x[:, 1], c=y, s=100, cmap='cool') #sample
plt.scatter(c[:, 0], c[:, 1], s=100, marker='^')
```

## Train model

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
k=5
clf= KNeighborsClassifier(n_neighbors=k)
clf.fit(x, y)
```

## predict

```python
x_sample = [0, 2]
x_sample2 = np.array(x_sample).reshape(1, -1)
y_sample = clf.predict(x_sample2)
neighbors = clf.kneighbors(x_sample2, return_distance=False)
```

```python
neighbors
```

## Plot the k neighbors

```python
plt.figure(figsize=(16,10), dpi=144)
plt.scatter(x[:, 0], x[:, 1], c=y, s=100, cmap='cool')
plt.scatter(x_sample[0], x_sample[1], marker='x', c=y_sample, s=100, cmap='cool')

for i in neighbors[0]:
    a, b = [x[i][0], x_sample[0]], [x[i][1], x_sample[1]]
    print a,b
    plt.plot(a, b, 'k--', linewidth=0.6)
```

## Regression

```python
n_dots = 40
x = 5 * np.random.rand(n_dots, 1)
y = np.cos(x).ravel()
y += 0.2 * np.random.rand(n_dots) - 0.1
```

```python
print x, y
```

```python
from sklearn.neighbors import KNeighborsRegressor
k = 5
knn = KNeighborsRegressor(k)
knn.fit(x, y)
```

```python
T = np.linspace(0, 5, 500)[:, np.newaxis]
print T
```

```python
y_pred = knn.predict(T)
knn.score(x, y)
```

```python
plt.figure(figsize=(16, 10), dpi=144)
plt.scatter(x, y, c='g', label='data', s=100)
plt.plot(T, y_pred, c='k', label='prediction', lw=4)
plt.axis('tight')
plt.title(r"KNeighborsRegressor (k= %i)" % k)
plt.show()
```
